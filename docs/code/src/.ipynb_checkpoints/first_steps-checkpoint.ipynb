# Read CSV
.. means we go out of the directory, each element of the path is then separated by a /

pd.read_csv has the following parameters:

- path: string containing the path to the file that we want to open (folders and files separated by /)
- sep: how each column in the file is separated.


```python
df = pd.read_csv("../data/materials.csv", sep=";")
df
```

```python
condition_1 = df["type"]=="Concrete"
condition_2 = df["embodied_carbon"] < 260
df.loc[condition_1 & condition_2]
```

```python
mask = []
for i in range(len(df)):
    remainder =  i % 2 
    isEven = remainder == 0
    if isEven:
        mask.append(True)
    else:
        mask.append(False)
        
df["id"] % 2 == 1
```

```python
rows = df["id"] % 2 == 1
columns = ["name", "type"]
df.loc[rows, columns]
```
